```python
# LeeML20 homework1 PM2.5 - regression
# 输入为9天所有数据 输出为第十天PM2.5
import numpy as np
import pandas as pd
```

```python
raw_train = pd.read_csv("train_.csv")
```

```python
raw_train.head()
```

```python
raw_train_dropdat = raw_train.drop('Date',axis=1)
raw_train_dropdat = raw_train_dropdat.set_index("Type")
```

```python
raw_train_dropdat
```

```python
#把各天各行拼接在1/1之后
new_train = raw_train_dropdat[0:18]
start = 18
end = 36
for i in range(1,240):
    tpt = raw_train_dropdat[start:end]
    new_train=pd.concat([new_train, tpt],axis=1)
    start += 18
    end += 18
# 切片后行序号会变 为了保证行序号重排 使用resetindex命令
```

```python
new_train = new_train.T.reset_index(drop=True,).T
```

```python
new_train.T.info()
# 下雨这一块应该把NR替换为0
```

```python
new_train_norain=new_train.replace('NR','0')
'NR' in set(new_train_norain.iloc[10])
```

```python
new_train_norain=new_train_norain.astype('double')
```

```python
new_train_norain.info()
```

```python
new_train_norain=new_train_norain.T.reset_index(drop=True).T
%matplotlib inline
import matplotlib.pyplot as plt
'''new_train_norain.T.hist(bins=50,figsize=(20,15))'''
```

```python
# 划分输入特征向量9天的所有数据与标签第十天PM2.5
# ！！！！注意！！！！测得一个月前20天的数据 不能一直遍历划分 需要跳过
# 似乎 一般的线性回归模型不需要标准化吗？？？
labelpoint = 9
feature = []
label = []
while(labelpoint<5760):
    startpoint = labelpoint - 9
    feature.append(new_train_norain.iloc[:,startpoint:labelpoint].values.reshape(1,-1))
    label.append(new_train_norain.iloc[9,labelpoint])
    if ((labelpoint+1)%480 == 0):
        labelpoint += 10
    else:
        labelpoint+=1
```

```python
new_feature = np.array([item[0] for item in feature])
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
scaler = StandardScaler()
new_feature=scaler.fit_transform(new_feature)
new_feature
```

```python
# 基于第一次学习后的过拟合现象，进行特征选取
# 先看静态指标
static_corr = new_train_norain.T.corr()
static_corr['PM2.5'].sort_values(ascending = False)
```

```python
new_df = pd.concat([pd.DataFrame(new_feature),pd.DataFrame(label).set_axis(['PM2.5'],axis=1)],axis=1)
```

```python
new_corr = new_df.corr()
corr_dict = dict(new_corr['PM2.5'].sort_values(ascending=False))
useless_feature=[(item,corr_dict[item]) for item in corr_dict if abs(corr_dict[item])<0]
print(len(useless_feature))
```

```python
df_feature = pd.DataFrame(new_feature)
for ele in useless_feature:
    df_feature.drop([ele[0]],axis=1,inplace=True)
df_feature
```

```python
tptlist=df_feature.values.tolist()
```

```python
X_train, X_test, y_train, y_test = train_test_split(tptlist, label, test_size=0.2, random_state=42)
np.random.seed(42)
weight = np.random.randn(162-len(useless_feature))
bias = float(np.random.rand(1))
weight=weight.reshape(-1,1)
X_train = np.array(X_train)
y_train = np.array(y_train).reshape(-1,1)
X_test = np.array(X_test)
y_test =np.array(y_test)
y_train.shape
# 多元线性回归需要标准化！！！
```

```python
'''epoch=100
LossList=[]
Lossval=[]
learning_rate = 0.02
# 损失函数用SSE L=∑(yi-y)^2
for loop in range(epoch):
    y_tpt = np.dot(X_train,weight) + bias
    y_tpt1 = np.dot(X_test,weight) + bias
    loss = 1/len(X_train)*np.sum((y_train - y_tpt)**2)
    Lossv = 1/len(X_test)*np.sum((y_test - y_tpt1)**2)
    Loss_to_w = -2 * np.dot(X_train.T, 1/len(X_train)*(y_train - y_tpt))
    Loss_to_b = np.sum( -2 * 1/len(X_train)*(y_train - y_tpt))
    bias = bias - learning_rate * Loss_to_b
    LossList.append(loss)
    Lossval.append(Lossv)'''
LossList=[]
dim = 18 * 9 + 1
w = np.zeros([dim, 1])
x = np.concatenate((np.ones([12 * 471, 1]), np.array(tptlist)), axis = 1).astype(float)
learning_rate = 100
iter_time = 800
adagrad = np.zeros([dim, 1])
eps = 0.0000000001
for t in range(iter_time):
    loss = np.sqrt(np.sum(np.power(np.dot(x, w) - np.array(label).reshape(-1,1), 2))/471/12)#rmse
    LossList.append(loss)
    if(t%100==0):
        print(str(t) + ":" + str(loss))
    gradient = 2 * np.dot(x.transpose(), np.dot(x, w) - np.array(label).reshape(-1,1)) #dim*1
    adagrad += gradient ** 2
    w = w - learning_rate * gradient / np.sqrt(adagrad + eps)
```

```python
test_data = pd.read_csv("test.csv", header=None)
```

```python
test_data=test_data.set_index([0],drop=True)
```

```python
test_data.drop([1],axis=1,inplace=True)
```

```python
test_data.replace('NR',0,inplace=True)
test_data=test_data.astype('double')
```

```python
test_data.iloc[0:18].values.reshape(1,-1)[0]
```

```python
test_feature=[]
startpoint = 0
endpoint = 18
for i in range(240):
    tptdf = test_data.iloc[startpoint:endpoint].values.reshape(1,-1)[0]
    test_feature.append(tptdf)
    startpoint+=18
    endpoint+=18
```

```python
tptarray = pd.DataFrame(test_feature)
```

```python
tptarray
```

```python
for ele in useless_feature:
    tptarray.drop([ele[0]],axis=1,inplace=True)
tptarray
```

```python
testlist=tptarray.values.tolist()
```

```python
testX = scaler.fit_transform(testlist)
```

```python
testX = np.concatenate((np.ones([240, 1]), np.array(testX)), axis = 1).astype(float)
testY = np.dot(testX,w)
```

```python
outY = pd.DataFrame(testY)
```

```python
indexx=['id_'+str(i) for i in range(240)]
dicxx={}
for i in range(240):
    if (testY[i]>0):
        dicxx[indexx[i]] = outY.iloc[i]
    else:
        dicxx[indexx[i]] = 0.5
outYY = pd.DataFrame(dicxx).T
outYY
```
